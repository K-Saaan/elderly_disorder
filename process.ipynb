```python
def create_dataframe_from_json_file(file_path):
    with open(file_path, 'r') as f:
        json_data = json.load(f)
    data = {}
    for i in tqdm(range(len(json_data))):
        profile_persona_id = json_data[i]['profile']['persona-id']
        human = ', '.join(json_data[i]['profile']['persona']['human'])
        computer = ', '.join(json_data[i]['profile']['persona']['computer'])
        type = json_data[i]['profile']['emotion']['type']
        situation = ', '.join(json_data[i]['profile']['emotion']['situation'])
        talk_id = json_data[i]['talk']['id']['talk-id']
        HS01 = json_data[i]['talk']['content']['HS01']
        SS01 = json_data[i]['talk']['content']['SS01']
        HS02 = json_data[i]['talk']['content']['HS02']
        SS02 = json_data[i]['talk']['content']['SS02']
        HS03 = json_data[i]['talk']['content']['HS03']
        SS03 = json_data[i]['talk']['content']['SS03']
        
        data[i] = {'profile_persona_id': profile_persona_id,
                'human': human,
                'computer': computer,
                'type': type,
                'situation': situation,
                'talk-id': talk_id,
                'HS01': HS01,
                'SS01': SS01,
                'HS02': HS02,
                'SS02': SS02,
                'HS03': HS03,
                'SS03': SS03}
    
    df = pd.DataFrame.from_dict(data, orient='index')
    return df
```

```python
# JSON 파일 경로
json_file_path = './data/Training/라벨링데이터/감성대화말뭉치(최종데이터)_Training.json'

# DataFrame 생성
df = create_dataframe_from_json_file(json_file_path)
df.head()
```

```python
def data_precess(df):
    df[['age', 'gender']] = df['human'].str.split(', ', expand=True)
    df[['situation01', 'situation02']] = df['situation'].str.split(', ', expand=True)
    df.drop(['human', 'situation'], axis=1, inplace=True)
    
    df.rename(columns={'age' : 'age_id',
                   'gender' : 'gender_id',
                   'situation01' : 'situation01_id',
                   'situation02' : 'situation02_id',
                   'type' : 'emotion_id'}, inplace=True)

    df['age'] = df['age_id'].apply(lambda x: age_dict[x])
    df['gender'] = df['gender_id'].apply(lambda x: gender_dict[x])
    df['situation01'] = df['situation01_id'].apply(lambda x: situation_dict[x])
    df['situation02'] = df['situation02_id'].apply(lambda x: disease_dict[x])
    df['emotion'] = df['emotion_id'].apply(lambda x: emotion_dict[x])
    
    return df
```
