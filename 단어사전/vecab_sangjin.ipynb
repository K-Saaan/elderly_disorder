```python
add_word = {
    '부정확' : 'NNG',
    '무질서' : 'NNG',
    '마감 기한' : 'NNG',
    '불안정' : 'NNG',
    '비정상' : 'NNG',
    '무분별' : 'NNG',
    '비지향' : 'NNG',
    '부주의' : 'NNG',
    '불승인' : 'NNG'
}
for k, v in add_word.items():
    kiwi.add_user_word(k, v)
```

```python
# 체언(NNG, NNP,  NR, NP), 용언(VX, VCN), 어근에 해당 하는 품사와 일반 부사(MAG), 체언 접두사(XPN)만을 선별
target_tags = ['NNG', 'NNP', 'NR', 'NP', 'VCN', 'MAG', 'XPN']
token_disorder = {}
for key in mental_disorder.keys():
    token_list = []
    for sentence in mental_disorder[key]:
        print(f'token : {sentence}')
        tokens = kiwi.tokenize(sentence)
        token_list.append([token.form for token in tokens if token.tag in target_tags])
    token_list = sum(token_list, [])
    token_disorder[key] = token_list
```

```python
remove_token = {'기타', '또는', '부', '곳', '전', '종종', '때', '무', '자주', '땅', '불', '모터', '과도', '줄', '그', '및', '이', '비', '산', '대', '부분', '즉', '유난히', '한', '그것', '길'}
filtered_token_disorder = {}

for disorder, tokens in token_disorder.items():
    filtered_tokens = [token for token in tokens if token not in remove_token]
    filtered_token_disorder[disorder] = filtered_tokens
```

```python
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import defaultdict

def preprocess_text(text):
    # 정규 표현식을 사용한 불필요한 문자 제거
    text = re.sub(r'\W', ' ', text)
    text = text.lower()  # 소문자 변환
    return text

def analyze_text(text, language='english'):
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stopwords.words(language)]
    # 형태소 분석 라이브러리를 사용한 추가 분석 가능
    return tokens

def process_diagnosis_criteria(criteria):
    processed_criteria = defaultdict(list)
    for disorder, symptoms in criteria.items():
        for symptom in symptoms:
            processed_text = preprocess_text(symptom)
            analyzed_tokens = analyze_text(processed_text)
            processed_criteria[disorder].append(analyzed_tokens)
    return processed_criteria

# 진단 기준 처리
processed_criteria = process_diagnosis_criteria(mental_disorder)
```
